# Manually Implementing a Simple Linear Regressor in Python

To manually implement a simple linear regressor in Python, we follow the basic linear regression formula:

$$ y = \beta_0 + \beta_1 \cdot x $$

Where:
- $ y $ is the predicted value,
- $ x $ is the feature (input data),
- $ \beta_0 $ is the intercept (bias),
- $ \beta_1 $ is the slope (weight).

We'll derive the coefficients $ \beta_0 $ and $ \beta_1 $ using the least squares method, and then implement this as code.

## Step-by-step breakdown:
1. **Calculate the mean** of both `X` (input) and `y` (output).
2. **Calculate the slope (β1)** using the formula:

   $$ \beta_1 = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sum{(x_i - \bar{x})^2}} $$

3. **Calculate the intercept (β0)** using the formula:

   $$ \beta_0 = \bar{y} - \beta_1 \cdot \bar{x} $$

4. **Make predictions** for new data using the formula $ y = \beta_0 + \beta_1 \cdot x $

```python
# Sample data (X: feature, y: target)
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 3.5, 5])
```

```python

# Step 1: Calculate the means of X and y
X_mean = np.mean(X)
y_mean = np.mean(y)
```

```python

# Step 2: Calculate the slope (β1)
numerator = np.sum((X - X_mean) * (y - y_mean))
denominator = np.sum((X - X_mean) ** 2)
beta_1 = numerator / denominator
```

```python

# Step 3: Calculate the intercept (β0)
beta_0 = y_mean - beta_1 * X_mean
```

```python

# Step 4: Define the prediction function (linear equation)
def predict(x):
    return beta_0 + beta_1 * x
```

```python

# Step 5: Make predictions for the input data
y_pred = predict(X)
```

```python

# Step 6: Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python

# Print the calculated parameters
print(f"Intercept (β0): {beta_0}")
print(f"Slope (β1): {beta_1}")
```
